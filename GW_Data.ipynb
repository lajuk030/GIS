<h2> Read Water Quality Data </h2>

```python
fname='WaterQualityMajor.txt'
watq= pd.read_csv(fname, sep='|', encoding= "ISO-8859-1", low_memory=False)
```

```python
watq.columns
```

```python
cols = ['StateWellNumber', 'SampleYear', 'ParameterCode', 'ParameterValue']
watqgc = watq.loc[watq['Aquifer']=='Gulf Coast',cols]
```

```python
watqgc.head()
```

```python
watqgc['ParameterCode']
```

```python
watqgc['EC']=watqgc.loc[watqgc['ParameterCode']==95, ['ParameterValue']]
```

```python
watqgc.columns
```

```python
len(watqgc)
```

```python
EC = watqgc.loc[watqgc['ParameterCode']==94, ['ParameterValue']]
len(EC)
EC
```

```python
watqgc['AS']=watqgc.loc[watqgc['ParameterCode']==1000, ['ParameterValue']]
```

```python
watqgc.columns
```

```python
len(watqgc)
```

```python
AS = watqgc.loc[watqgc['ParameterCode']==1000, ['ParameterValue']]
len(AS)
AS
```

```python
watqgc
```

```python
watqcgrp = watqgc.groupby('StateWellNumber')[['StateWellNumber', 'EC', 'AS']].mean()
watqcgrp = watqcgrp.reset_index(drop=True)
print(watqcgrp)
```

```python
watqcLL = pd.merge(GCdat, watqcgrp, on = 'StateWellNumber', how = 'inner')
watqcLL
```

```python
watqECASp = watqcLL.dropna(subset = ['EC','AS'], how='all')[['StateWellNumber', 'LatitudeDD', 'LongitudeDD', 'EC', 'AS']]
```

```python
watqECASp.head()
```

```python
len(watqECASp)
```

```python
watqECASp.to_csv('GCESAS.csv', index=False)
```
