```python
#Conver CRS Values
texasGCS84 = texas.to_crs(epsg=4326)
texasGCS84.head()
```

```python
texasNAD27 = texas.to_crs(epsg=4267)
texasNAD27.head()
```

```python
texasGCS84.plot()
```

```python
texasNAD27.plot()
```

```python
texasUTM14N = texasGCS84.to_crs(epsg=32614)
texasUTM15N = texasGCS84.to_crs(epsg=32615)
```

```python
#CF is sqm to sqmi
CF=3.861021585E-07
texasUTM14N['Areasqmi'] = texasUTM14N.area*CF
texasUTM15N['Areasqmi'] = texasUTM15N.area*CF
```

```python
#np.float64
texasUTM14N.Areasqmi.mean()
```

```python
texasUTM15N.Areasqmi.mean()
```

```python
texasAEC = texasGCS84.to_crs(epsg=5070)
```

```python
texasAEC.plot()
```

```python
texasAEC['Areasqmi'] = texasAEC.area*CF
```

```python
texasAEC.Areasqmi.mean()
```

```python
AECmean = texasAEC.Areasqmi.mean()
```

```python
AECmean = float(AECmean)         # Convert to float
round(AECmean, 3)     # Round to 3 decimal places and reassign
```

```python
#Convert AEC to WGS84 and plot and color county by size
texasxWGS84a = texasAEC.to_crs(epsg=4326)
texasxWGS84a.plot(column = 'Areasqmi',cmap='viridis',legend=True, edgecolor='black')
```

```python
texasxWGS84a.plot(column = 'Areasqmi',cmap='inferno',legend=True, edgecolor='black')
```

```python
texasxWGS84a.plot(column = 'Areasqmi',cmap='hsv',legend=True, edgecolor='black')
```
